# Loyalty lost

## 2014

### Members who joined on Christmas: **2**
### Members who joined on New Year: **4**

## 2015

### Members who joined on Christmas: **0**
### Members who joined on New Year: **2**

```python
%matplotlib inline
import pandas as pd
import seaborn 

pd.DataFrame(data={'2014': [2,4], '2015': [0,2]}, index=['Christmas', 'New Year']).plot(kind='bar', title='Loyalty Lost')
```
