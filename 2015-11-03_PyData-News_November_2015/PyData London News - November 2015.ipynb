# PEP 492 - Coroutines with async and await syntax

Coroutine functions are declared using the new async def syntax:

```python
async def coro():
    return 'spam'
```

Inside a coroutine function, the new await expression can be used to suspend coroutine execution until the result is available. Any object can be awaited, as long as it implements the awaitable protocol by defining the `__await__()` method.

# PEP 465 - A dedicated infix operator for matrix multiplication

Matrix mutplication operator, `@`, for cleaner code!

```python
import numpy as np

x = np.ones(3)
m = np.eye(3)

x @ m
```

* `__matmul__()`, regular
* `__rmatmul__()`, reflected
* `__imatmul__()`, in-place

# PEP 471 - os.scandir() function  a better and faster directory iterator

* 3 to 5 times faster on POSIX systems and 7 to 20 times faster on Windows systems
* Speed-up largely due reduction of `os.stat()` calls and returns a generator 
* `os.walk()` now takes advantage of `os.scandir()`

```python
for entry in os.scandir(path):
    if not entry.name.startswith('.') and entry.is_file():
        print(entry.name)
```

# POTM, DDT - multiply one test case by running it with different test data

* A collection of decorators to throw more data at tests
* Can read data from file! `@file_data`

```python
@file_data('testdata.json')
def test_larger_than_two(self, value):
    self.assertTrue(larger_than_two(value))
```
